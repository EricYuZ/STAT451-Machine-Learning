```python
fig = plt.figure() # create new blank figure
fig.suptitle('gridspec: three plots using 1/4, 1/4, and 1/2 of figure')
gs = fig.add_gridspec(nrows=4, ncols=1) # grid of plot axes
x = np.linspace(start=0, stop=4*np.pi, num=100) # fake data
xlim = (0, 13) # common x-axis limits

ax1 = fig.add_subplot(gs[0, 0]) # first plot in top 1/4
ax1.plot(x, -2 + x/3, label=r'$y = -2 + \frac{x}{3}$')
ax1.set(xlim=xlim, ylim=(-2, 2))
ax1.xaxis.set_ticks([]) # suppress axis ticks & numbers
ax1.legend()

ax2 = fig.add_subplot(gs[1, 0]) # second plot in second 1/4
ax2.plot(x, np.sin(x), label=r'$y = \cos(x)$')
ax2.set(xlim=xlim, ylim=(-2, 2))
ax2.xaxis.set_ticks([]) # suppress axis ticks & numbers
ax2.legend()

ax3 = fig.add_subplot(gs[2:4, 0]) # third plot in bottom 1/2
ax3.plot(x, (-2 + x/3) + np.sin(x),
         label=r'$y=\left(-2 + \frac{x}{3}\right) + \cos(x)$')
ax3.set(xlim=xlim, ylim=(-2, 6))
ax3.legend()
```

```python
import numpy as np # for np.math.factorial
def baby_binom_pmf(x, n, p):
    prob = np.math.factorial(n)/(np.math.factorial(x)*np.math.factorial(n-x))*p**x*(1-p)**(n-x)
    return prob
    
    
assert np.isclose(.25, baby_binom_pmf(0, 2, .5))
assert np.isclose(.50, baby_binom_pmf(1, 2, .5))
assert np.isclose(.25, baby_binom_pmf(2, 2, .5))
```

```python
def is_lower_vowel(letter):
    """Returns whether letter is a lower-case vowel (a, e, i, o, u)."""
    if letter == 'a' or letter == 'e' or letter == 'i' or letter == 'o' or letter == 'u':
        return True
    else:
        return False
```

```python
def baby_min(x, y):
    """Returns the smaller of x and y (both scalars)."""
    if x>y:
        return y
    else:
        return x
```

```python
def age_category(age):
    """Given age, returns 'child', 'youth', 'adult' or 'senior'.
    More precisely, given a nonnegative integer age, returns a string,
    one of 'child' (0-14 years), 'youth' (15-24 years), 'adult' (25-64 years),
    or 'senior' (65 or older).
    """
    if age > 64:
        return('senior')
    elif age > 24:
        return('adult')
    elif age > 14:
        return('youth')
    else:
        return ('child')
    
```

```python
def is_triangle(a, b, c):
    """Returns whether lengths a, b, and c can form a triangle.
   
    More specifically, returns whether a, b, and c are positive
    with the sum of any two greater than the third.
    """
    if (a+b> c and a+c>b and b+c>a)and (a>0 and b>0 and c>0):
        return True
    else:
        return False
```
